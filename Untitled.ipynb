```python
import re
import pandas as pd
```

```python
f = open("_chat.txt",'r', encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

#### Reg expression  -> regex 101.com 

```python
pattern=  '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}:\d{2}\s[a-zA-Z\s]M]'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_messages': messages, 'message_date': dates})
# Convert message data type =
df['message_date']= pd.to_datetime(df['message_date'],format='%d/%m/%y, %H:%M:%S %p]')


df.rename(columns={'message_date': 'date'},inplace=True)

df.head()
```

### Separate user_name from message

```python
users = []
messages =[]
for message in df['user_messages']:
    entry = re.split ('([\w\W]+?):\s', message)
    if entry[1:]:# user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['user'] = users
df['message'] =messages
df.drop(columns=['user_messages'], inplace=True)
df.all()
    
```

```python
df['month']= df['date'].dt.month_name()
df['day']= df['date'].dt.day
df['hour']= df['date'].dt.hour
df['minute']= df['date'].dt.minute
```
